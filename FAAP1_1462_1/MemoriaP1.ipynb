### Funciones de validación con sklearn

```python
def ValidacionSimpleSK(datos, porcentaje):
    data = dataset.datos[:,:-1]
    clases = dataset.datos[:,-1] 
    
    return train_test_split(data, clases, train_size = porcentaje, test_size = 1 - porcentaje, shuffle=True)
```

```python
def ValidacionCruzadaSK(datos, k):
    data = dataset.datos[:, :-1]
    clases = dataset.datos[:, -1]
        
    particiones = []

    kf = KFold(n_splits=k, shuffle=True)
    for train, test in kf.split(data, clases):
        particiones.append(EstrategiaParticionado.Particion(train, test))

    return particiones
```

### Clasificación Naive Bayes

```python
class NaiveBayesSKMult(Clasificador.Clasificador):
    
    def __init__(self):
        super().__init__()
        self.NB = nb.MultinomialNB(alpha=1.0, fit_prior=True, class_prior=None)
    
    def entrenamiento_datos_directos(self, dataTrain, clasesTrain):
        self.NB.fit(dataTrain, clasesTrain)
    
    def entrenamiento_indices(self, datos, indices):
        self.NB.fit(datos[indices][:,:-1].astype(np.float), datos[indices][:,-1].astype(np.float))
        
    def clasifica_datos_directos(self, dataTrain):
        return self.NB.predict(dataTrain)   

    def clasifica_indices(self, datos, indices):
        return self.NB.predict(datos[indices][:,:-1].astype(np.float))
```

```python
class NaiveBayesSKGauss(Clasificador.Clasificador):
    
    def __init__(self):
        super().__init__()
        self.NB = nb.GaussianNB()
    
    def entrenamiento_datos_directos(self, dataTrain, clasesTrain):
        self.NB.fit(dataTrain, clasesTrain)
    
    def entrenamiento_indices(self, datos, indices):
        self.NB.fit(datos[indices][:,:-1].astype(np.float), datos[indices][:,-1].astype(np.float))
        
    def clasifica_datos_directos(self, dataTrain):
        return self.NB.predict(dataTrain)   

    def clasifica_indices(self, datos, indices):
        return self.NB.predict(datos[indices][:,:-1].astype(np.float))
```

```python
def ScikitError(datos,pred):
		# Aqui se compara la prediccion (pred) con las clases reales y se calcula el error
		return 1 - np.sum(np.equal(datos.astype(int), pred.astype(int)))/len(pred)
```

#### Validación simple (german.data)

```python
dataset = Datos.Datos('ConjuntoDatos/german.data')
dataTrain, dataTest, clasesTrain, clasesTest = ValidacionSimpleSK(dataset,0.6)
clasificador = NaiveBayesSKGauss()
errores = []

clasificador.entrenamiento_datos_directos(dataTrain.astype(np.float), clasesTrain.astype(np.float))

pred = clasificador.clasifica_datos_directos(dataTest.astype(np.float))
errores.append(ScikitError(clasesTest, pred))
print("proporción error: ", errores)
mediaVSG = errores[0]
stdVSG = 0.0
```

#### Validación cruzada (german.data)

```python
dataset = Datos.Datos('ConjuntoDatos/german.data')
particiones = ValidacionCruzadaSK(dataset, 4)
clasificador = NaiveBayesSKGauss()
errores = []
for particion in particiones:
    clasificador.entrenamiento_indices(dataset.datos, particion.indicesTrain)
    pred = clasificador.clasifica_indices(dataset.datos, particion.indicesTest)
    clasesTest = dataset.datos[particion.indicesTest][:,-1].astype(np.float)
    errores.append(ScikitError(clasesTest, pred))

mean = np.mean(errores)
std = np.std(errores)
print("media: ", mean)
print("desviación: ", std)
mediaVCG = mean
stdVCG = std
```

### Clasificador NB en el espacio ROC para german

```python
dataset = Datos.Datos('ConjuntoDatos/german.data')
model = nb.GaussianNB()
trainX, testX, trainy, testy = ValidacionSimpleSK(dataset.datos, 0.75)
model.fit(trainX.astype(np.float), trainy.astype(np.float))

ns_probs = [0 for _ in range(len(testy))]
# predict probabilities
lr_probs = model.predict_proba(testX.astype(np.float))
# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
# calculate scores
arraygood = []
for num in testy:
    arraygood.append(int(num) - 1)
    
ns_auc = roc_auc_score(testy, ns_probs)
lr_auc = roc_auc_score(testy, lr_probs)
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Naive Bayes: ROC AUC=%.3f' % (lr_auc))
ns_fpr, ns_tpr, _ = roc_curve(arraygood, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(arraygood, lr_probs)
pyplot.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
pyplot.plot(lr_fpr, lr_tpr, marker='.', label='Naive Bayes')
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
pyplot.legend()
pyplot.show()
```

Obtenemos una AUC de valor 0.765, utilizando validación simple
